```python
%matplotlib inline

import math
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd

np.random.seed(1)

import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
from tensorflow.python.framework import ops
from sklearn.model_selection import train_test_split
from pandas.util.testing import assert_frame_equal
```

```python
train = pd.read_csv("fashion-mnist_train_small.csv")
test = pd.read_csv("fashion-mnist_test_small.csv")
#train, cross_val = train_test_split(full_train, test_size=0.2)
```

```python
train_np = train.as_matrix()
train_y = train_np[:,0]
train_x = train_np[:, 1:]
```

```python
model = Sequential()
model.add(Dense(units=70, use_bias=True, bias_initializer='RandomUniform', kernel_initializer='RandomUniform', 
                activation='relu', input_dim=train_x.shape[1]))
model.add(Dense(units=50, use_bias=True, bias_initializer='RandomUniform', kernel_initializer='RandomUniform', 
                activation='relu'))
model.add(Dense(units=30, use_bias=True, bias_initializer='RandomUniform', kernel_initializer='RandomUniform', 
                activation='relu'))
model.add(Dense(units=10, use_bias=True, bias_initializer='RandomUniform', kernel_initializer='RandomUniform', 
                activation='sigmoid'))

model.compile(optimizer=keras.optimizers.SGD(), loss='categorical_crossentropy', metrics=['accuracy'])

target = to_categorical(train_y)
history = model.fit(train_x, target, epochs=350, batch_size=50)
#scores = model.evaluate(train_x, train_y)
#print('train accuracy: ' + scores[1])
```

```python
test_np = test.as_matrix()
test_y = test_np[:, 0]
test_x = test_np[:, 1:]
y_pred = model.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y, np.argmax(y_pred, axis = 1).reshape(-1, 1))
```

```python
plt.plot(history.history['acc'])
plt.title('train accuracy learning curve')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title('train loss learning curve')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
